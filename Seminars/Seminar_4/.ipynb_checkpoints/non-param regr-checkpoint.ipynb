## Ядерные оценки плотности:

```python
import statsmodels as sm
nobs = 300
np.random.seed(1234)  # Seed random generator
plt.figure(figsize = (12, 8))
dens = sm.nonparametric.kde.KDEUnivariate(np.random.normal(size=nobs))
for bw in [0.01, 0.1, 0.2, 0.3, 0.5, 1, 2]:
    dens.fit(bw = bw)
    plt.plot(np.linspace(-4,4), dens.evaluate(np.linspace(-4,4)), label = 'bw = {}'.format(bw))
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
ran = np.linspace(-5,5)
nobs = 5
sample = np.random.normal(size=nobs)
dens2 = sm.nonparametric.kde.KDEUnivariate(sample)
for kernel in[
'epa',
'gau',
'tri',
'uni']:
    dens2.fit(kernel = kernel, fft = False)
    plt.plot(ran, [dens2.evaluate(x) for x in ran], label = 'kernel = {}'.format(kernel))
plt.scatter(sample, np.repeat(0, nobs))
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
ran = np.linspace(-5,5)
nobs = 60
sample = np.random.normal(size=nobs)
dens2 = sm.nonparametric.kde.KDEUnivariate(sample)
for kernel in[
'epa',
'gau',
'tri',
'uni']:
    dens2.fit(kernel = kernel, fft = False)
    plt.plot(ran, [dens2.evaluate(x) for x in ran], label = 'kernel = {}'.format(kernel))
plt.scatter(sample, np.repeat(0, nobs))
plt.legend()
plt.show()
```

## Непараметрическая регрессия:

```python
x = np.random.randn(20)*2
n = len(x)
y = 6*x - x**3 + 3*np.random.randn(n)
ran = np.linspace(x.min(), x.max())
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as MSE
```

```python
lr = LR()
lr.fit(x.reshape(-1, 1), y)

plt.scatter(x,y)
plt.plot(ran, lr.predict(ran.reshape(-1, 1)), color = 'red')
```

```python
KR = KernelRegression()
KR.fit(x.reshape(-1, 1), y)

plt.scatter(x,y)
plt.plot(np.linspace(-4,4), KR.predict(np.linspace(-4,4).reshape(-1, 1)), color = 'red')
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x,y, label = 'True')
for gam in [0.1, 1, 5, 50]:
    KR = KernelRegression(gamma = gam)
    KR.fit(x.reshape(-1, 1), y)
    plt.plot(ran, KR.predict(ran.reshape(-1, 1)), label = 'gamma = {}'.format(gam))
plt.legend()
```

```python
s1 = np.arange(20)
s2 = 0.1*s1 + 0.01*s1**2 + np.random.randn(20)
```
